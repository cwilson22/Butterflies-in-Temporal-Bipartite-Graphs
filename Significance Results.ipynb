# Significance Results versus Inter-Event Shuffling

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(1, 3, figsize=(20, 8))
fig.suptitle("Significance Figures versus Inter-Event Shuffling RRMs", fontsize=30)

data = pd.read_csv("notebookdata/8x(t)sigdata", index_col=0)
sns.heatmap(data, ax=axs[0], square=True, cmap=sns.color_palette('bwr', as_cmap=True), center=0)
axs[0].set_title("8*x(dt)", fontsize=20)

data = pd.read_csv("notebookdata/32x(t)sigdata", index_col=0)
sns.heatmap(data, ax=axs[1], square=True, cmap=sns.color_palette('bwr', as_cmap=True), center=0)
axs[1].set_title("32*x(dt)", fontsize=20)

data = pd.read_csv("notebookdata/128x(t)sigdata", index_col=0)
sns.heatmap(data, ax=axs[2], square=True, cmap=sns.color_palette('bwr', as_cmap=True), center=0)
axs[2].set_title("128*x(dt)", fontsize=20)
plt.show()
```

### Significance figures refers to normalized z-scores (normalized with respect to the z-scores for the other five types of butterflies) of the actual butterfly counts against 10 repetitions of the inter-event shuffling randomized reference model.

### Results are given for $\Delta_C = 8x(\Delta t), 32x(\Delta t),$ and $128x(\Delta t)$, for 10 different networks.
